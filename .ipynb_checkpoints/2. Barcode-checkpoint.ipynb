```python
import cv2
import glob
from pyzbar import pyzbar
import matplotlib.pyplot as plt
from pprint import pprint
```

```python
def read_barcode(img):
    
    barcodes = pyzbar.decode(img)
    barcodes_data = []
    for barcode in barcodes:
        barcode_info = {}
        barcode_info['code'] = barcode.data.decode('utf-8')
        left, top, width, height = barcode.rect
        barcode_info['rec'] = {'left':left,'top':top,'width':width,'height':height}
        barcode_info['polygon'] = [{'x':p.x, 'y':p.y} for p in barcode.polygon]
        barcodes_data.append(barcode_info)
    return {'data' : barcodes_data}

def draw_barcode(img, barcode_info): 
    data = barcode_info['data']
    
    for barcode in data:
        x,y,w,h = barcode['rec']['left'], barcode['rec']['top'], \
                  barcode['rec']['width'], barcode['rec']['height']
        
        cv2.rectangle(img, (x, y),(x+w, y+h), (0, 255, 0), 2)
        font = cv2.FONT_HERSHEY_DUPLEX
        cv2.putText(img, barcode['code'], (x + 6, y - 6), font, 0.6, (189, 230, 55), 1)
    return img
```

```python
img = cv2.imread('barcodes/23.png')
info = read_barcode(img)
plt.imshow(draw_barcode(img,info))
plt.show()
pprint(info)
```

```python
img_paths = glob.glob('barcodes/*')
for img_path in img_paths:
    img = cv2.imread(img_path)
    info = read_barcode(img)
    plt.imshow(draw_barcode(img,info))
    plt.show()
    pprint(info)
```
