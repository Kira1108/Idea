```python
from cnstd import CnStd
from cnocr import CnOcr
import cv2
import matplotlib.pyplot as plt
import glob
```

```python
IMAGE_PATHS = glob.glob("ocr_imgs/*")
```

```python
std = CnStd()
cn_ocr = CnOcr()
```

### 场景文字分割

```python
def read_img(path):
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img

def detect_image(img):
    return std.detect(img)

def plot_crops(img, result):
    n_crops = len(result)
    cols = 5
    rows = n_crops // cols + int(n_crops % 5 > 0)
    
    plt.subplots(figsize = (15,5))
    for i,box in enumerate(result):
        plt.subplot(rows, cols, i+1)
        b = box['box']
        xmin, xmax = b[:,0].min(),b[:,0].max()
        ymin, ymax = b[:,1].min(),b[:,1].max()
        plt.imshow(img[ymin:ymax, xmin:xmax])
    plt.show()

def test():
    for p in IMAGE_PATHS:
        try:
            img = read_img(p)
            plt.imshow(img)
            result = detect_image(img)
            plot_crops(img, result)
        except:
            print('Error on img {}'.format(p))
```

```python
test()
```

```python
results = []
imgs = []
for p in IMAGE_PATHS:
    try:
        img = read_img(p)
        plt.imshow(img)
        result = detect_image(img)
        results.append(result)
        imgs.append(img)
    except:
        print('Error on img {}'.format(p))
```

```python
for i, r in enumerate(results):
    img = imgs[i]
    for box in r:
        try:
            b = box['box']
            xmin, xmax = b[:,0].min(),b[:,0].max()
            ymin, ymax = b[:,1].min(),b[:,1].max()
            cropped_img = img[ymin:ymax, xmin:xmax]
            texts = cn_ocr.ocr_for_single_line(cropped_img)
            plt.imshow(cropped_img)
            print(texts)
            plt.show()
        except:
            pass
```
