```python
from cnstd import CnStd
from cnocr import CnOcr
import cv2
import matplotlib.pyplot as plt
import glob
```

```python
IMAGE_PATHS = glob.glob("ocr_imgs/*")
```

```python
std = CnStd()
cn_ocr = CnOcr()
```

```python
img_path = "ocr_imgs/6253aebeb25a9d5913e1e2894de1145c.jpeg"
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
result = std.detect(img)
```

```python
cols = 5
rows = len(result) // cols + int(len(result) % cols > 0)

fig, ax = plt.subplots(figsize = (20,8))
for i, box in enumerate(result):
    plt.subplot(rows, cols, i + 1)
    plt.imshow(box['cropped_img'])
plt.show()
```
